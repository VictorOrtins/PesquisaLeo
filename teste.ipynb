```python
import tensorflow as tf
import cv2
import numpy as np
```

```python
new_model = tf.keras.models.load_model('modelo.h5')
```

```python
new_model.summary()
```

```python
img = cv2.imread('(9961)2_R_4_3.png', cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (272, 360))  # Redimensionar para as dimensões esperadas pelo modelo
img = img / 255.0  # Normalizar os valores dos pixels para o intervalo [0, 1]
img = img[np.newaxis, ..., np.newaxis]

teste = new_model.predict(img)
print(teste)
```

```python
prediction = teste[0, ..., 0]

# # Converter para imagem binária usando um limiar (threshold)
# threshold = 0.99  # Ajuste conforme necessário
# binary_image = (prediction > threshold).astype(np.uint8)

# Escalonar de volta para o intervalo [0, 255]
binary_image = (prediction* 255).astype(np.uint8)
cv2.imwrite(f'teste.png', binary_image)

```
